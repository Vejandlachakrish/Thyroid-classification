# hybrid model-1

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# Define base models
decision_tree = DecisionTreeClassifier(random_state=0)
svm = SVC(probability=True, random_state=0)
logistic_regression = LogisticRegression(random_state=0)
```

```python
# Create a hybrid model with VotingClassifier
hybrid_model = VotingClassifier(
    estimators=[('dt', decision_tree), ('svm', svm), ('lr', logistic_regression)],
    voting='soft'  # Use soft voting to take probability into account
)
```

```python
# Train the hybrid model
hybrid_model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = hybrid_model.predict(x_test)
```

```python
# Evaluate the hybrid model
accuracy = accuracy_score(y_test, y_pred)
print(f"Hybrid Model Accuracy: {accuracy:.2f}")
```

```python
hybrid1_con=confusion_matrix(y_test,y_pred)
hybrid1_con
```

```python
hybrid1_cr=classification_report(y_test,y_pred)
hybrid1_cr
```

# hybrid model-2

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import StackingClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# Define base models
base_models = [
    ('decision_tree', DecisionTreeClassifier(random_state=42)),
    ('knn', KNeighborsClassifier(n_neighbors=5)),
    ('svm', SVC(probability=True, random_state=42))
]
```

```python
# Define the meta-model
meta_model = LogisticRegression()
```

```python
# Create a stacked hybrid model
stacked_model = StackingClassifier(estimators=base_models, final_estimator=meta_model)
```

```python
# Train the stacked model
stacked_model.fit(x_train, y_train)
```

```python
# Make predictions
y_pred = stacked_model.predict(x_test)
```

```python
# Evaluate the stacked model
accuracy = accuracy_score(y_test, y_pred)
print(f"Stacked Model Accuracy: {accuracy:.2f}")
```

```python
hybrid2_con=confusion_matrix(y_test,y_pred)
hybrid2_con
```

```python
hybrid2_cr=classification_report(y_test,y_pred)
hybrid2_cr
```
